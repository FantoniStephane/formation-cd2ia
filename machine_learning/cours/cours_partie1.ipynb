## 🧠 3. Concepts Fondamentaux

### 3.1 La hiérarchie IA / ML / Deep Learning



Ces trois termes sont souvent confondus mais désignent des niveaux différents :

| Terme | Définition |
|-------|------------|
| **IA** | Ensemble des systèmes simulant l'intelligence humaine |
| **Machine Learning** | Sous-partie de l'IA — apprendre à partir de données |
| **Deep Learning** | Sous-partie du ML — utilise des réseaux de neurones profonds |

> 💡 Le Deep Learning est inclus dans le ML, qui est inclus dans l'IA.
> C'est comme des poupées russes !


### 3.2 Composantes d'un modèle



Tout modèle ML est composé de :
- **Paramètres (poids)** : les valeurs que le modèle apprend
- **Fonction d'erreur (Loss Function)** : mesure l'écart entre prédiction et réalité
- **Hyperparamètres** : les réglages fixés avant l'entraînement

### 3.3 Les 3 méthodes d'apprentissage



| Méthode | Principe | Exemple |
|---------|----------|---------|
| **Supervisé** | Dataset annoté — un point / un label | Classification chat/chien |
| **Non Supervisé** | Dataset non annoté — trouver des groupes | Clustering, Régression |
| **Par Renforcement** | Agent qui apprend par récompenses | Jeux vidéo, robotique |


## 🔧 4. Principaux Modèles Supervisés

### 4.1 Régression Linéaire


Calcule la **droite** qui minimise l'écart à l'ensemble des points.
Variante : Régression Polynomiale (courbe de degré N au lieu d'une droite).

### 4.2 Support Vector Machines (SVM)


Trouve le **plan séparant le mieux** les classes.
- **Linéaire** : séparation directe
- **Non linéaire (Kernel Trick)** : transforme les données dans un espace de dimension supérieure

### 4.3 Régression Logistique


Utilise une **fonction sigmoïde** pour classifier — modèle probabiliste.
On définit un **seuil (threshold)** pour décider de la classe.


### 4.4 Classification Naïve Bayésienne


Calcule la **probabilité d'appartenance** à chaque classe selon les features.
Suppose l'indépendance des variables — modèle simple et rapide.
> Formule de Bayes : **P(Ck|X) = P(X|Ck) × P(Ck) / P(X)**

### 4.5 Arbres de Décision


Ensemble de **règles sous forme de nœuds et feuilles**.
- Classification : valeurs discrètes
- Régression : valeurs continues
- Combinaison d'arbres → **Decision Forests (Random Forest)**


### 4.6 Réseaux Neuronaux


Combinaison non linéaire de poids et paramètres organisés en couches.
- Chaque neurone : `entrée = w1×x1 + ... + wn×xn`
- Chaque neurone applique une **fonction d'activation** : `sortie = f(entrée)`


## 📊 5. Métriques de Mesure

### 5.1 La matrice de confusion



| | Prédit Positif | Prédit Négatif |
|-|----------------|----------------|
| **Réel Positif** | TP (Vrai Positif) | FN (Faux Négatif) |
| **Réel Négatif** | FP (Faux Positif) | TN (Vrai Négatif) |

### 5.2 Les 4 métriques principales

Exemple : 100 images, 25 oiseaux, 75 autres. Le modèle classe 20 oiseaux correctement + 30 autres comme oiseaux.

| Métrique | Formule | Résultat exemple | Utilité |
|----------|---------|------------------|---------|
| **Accuracy** | (TP + TN) / Total | 20/100 = **20%** | Vue générale si dataset équilibré |
| **Précision** | TP / (TP + FP) | 20/50 = **40%** | Quand on veut éviter les faux positifs |
| **Rappel** | TP / (TP + FN) | 20/25 = **80%** | Quand on veut détecter tous les positifs |
| **F1-Score** | TP / (TP + ½(FP+FN)) | — | Dataset déséquilibré |

> ⚠️ L'Accuracy peut être **trompeuse** sur un dataset déséquilibré !
> Exemple : 1% positif, 99% négatif → Accuracy 99% en prédisant toujours négatif,
> mais F1-Score = **0** — le modèle ne sert à rien !


## 🎯 6. Choix du Modèle

### 6.1 Nature des données


- **Type** : texte, valeurs discrètes/continues, images...
- **Dimension** : nombre de features
- **Répartition** : linéaire, polynomiale...
- **Équilibrage** : classes bien réparties ou déséquilibrées ?

### 6.2 Compromis Biais / Variance



| Problème | Cause | Symptôme |
|----------|-------|----------|
| **Underfitting** | Biais trop élevé | Modèle trop simple — rate les tendances |
| **Overfitting** | Variance trop élevée | Modèle trop complexe — mauvaise généralisation |

> 💡 Le bon modèle est celui qui **minimise les deux** !

### 6.3 Démarche générale


1. **Comprendre les données** — analyse, corrélations, visualisation
2. **Anticiper les corrélations** — quel modèle semble cohérent ?
3. **Tester les modèles** — entraîner et évaluer
4. **Interpréter et adapter** — changer de modèle ou régler les paramètres
