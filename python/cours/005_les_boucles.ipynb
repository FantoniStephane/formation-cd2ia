# Les boucles (while, while True, for, itérables)

Ce notebook couvre en profondeur les **boucles** en Python : `while`, `while True` avec `break/continue`, `for` sur différents **itérables** (listes, tuples, dictionnaires, ensembles, chaînes), `range`, `enumerate`, `zip`, la clause `else` sur les boucles, et les **bonnes pratiques**. Vous trouverez de nombreux **exemples exécutables** et une série d’**exercices** avec corrigés.

## 1. Itérables vs Itérateurs

**Définition :**

* **Itérable** :
  C'est **un objet que l'on peut parcourir élément par élément**.
  → Exemple : une **liste**, une **chaîne de caractères**, etc.

  👉 On peut le donner à une boucle `for`, ou le transformer en itérateur avec `iter(obj)`.

* **Itérateur** :
  C'est **l'outil qui parcourt réellement** l'itérable.
  Il **mémorise où on en est** dans le parcours, et on récupère chaque élément grâce à la fonction `next()`.


**🧠 On peut retenir :**

> Un **itérable**, c'est comme **un livre**.
> Un **itérateur**, c'est **le marque-page** qui avance page après page.


`````python
# Notre itérable (une liste)
items = [10, 20, 30]

# On crée un itérateur à partir de cette liste
it = iter(items)

print(next(it))  # 10 → première page
print(next(it))  # 20 → on avance
print(next(it))  # 30 → dernière page

# Si on continue, il n'y a plus de page :
# print(next(it))  # StopIteration
`````

💡 **En résumé :**

| Terme            | Rôle                                                              | Exemple                                |
| ---------------- | ----------------------------------------------------------------- | -------------------------------------- |
| **Itérable**     | Contient plusieurs éléments qu'on peut parcourir                  | `[10, 20, 30]`, `"Python"`, `range(5)` |
| **Itérateur**    | Produit les éléments un à un quand on l'appelle avec `next()`     | résultat de `iter([10, 20, 30])`       |
| **Boucle `for`** | Se charge automatiquement de créer l'itérateur et de le parcourir | `for x in [10, 20, 30]: ...`     

## 2. Boucle `while`

La boucle `while` répète un bloc **tant que** la condition est vraie.

⚠️ Attention au **risque de boucle infinie** si la condition ne devient jamais fausse.

`````python
# Compter jusqu'à 5
i = 1

while i <= 5:
    print(i)
    i = i + 1

print("\nTerminé.") # Note : ici, le carcatère "\n" représente un saut de ligne.
`````

## 3. `while True` avec `break` / `continue`

`while True` crée une boucle potentiellement infinie ; on en sort avec `break`. `continue` passe directement à l'itération suivante.

`````python
while True:
    mot_de_passe = input("Entrez votre mot de passe : ")

    if mot_de_passe == "":
        print("Mot de passe vide, réessayez.")
        continue  # on revient au début de l'itération (ligne 2)

    if mot_de_passe == "python123":
        print("Mot de passe correct !")
        break
    else:
        print("Mot de passe incorrect, essayez encore.")

print("Vous êtes en ligne !")
`````

## 4. Boucle `for` sur différents itérables


La boucle `for` itère directement sur les **éléments** d'un itérable.

### 4.1 Listes

`````python
fruits = ["pomme", "banane", "cerise"]
for f in fruits:
    print(f)
`````

### 4.2 Tuples (itération et déballage)

`````python
points = [
  (1, 2, 5),
  (3, 4, 9),
  (5, 6, 54)
]

for a, b, c in points:  # unpacking (ou déballage)
    print(a + b + c)
`````

### 4.3 Ensembles (`set`) non ordonnés

`````python
technos = {"python", "javascript", "go"}
for t in technos:
    print(t)
# L'ordre d'affichage peut varier (set non ordonné)
`````

### 4.4 Dictionnaires : `keys()`, `values()`, `items()`

`````python
user = {"nom": "Ada", "prenom": "Lovelace", "annee": 1815}

for cle in user.keys():
    print("clé:", cle)

for val in user.values():
    print("valeur:", val)

for cle, val in user.items():
    print(f"{cle} -> {val}")
`````

### 4.5 Chaînes de caractères

`````python
for lettre in "Hello":
    print(f">>>>> {lettre}")
`````

## 5. `range`, `enumerate`, `zip`

- `range(stop)` ou `range(start, stop[, step])` génère une progression d'entiers.
- `enumerate(itérable, start=0)` donne des couples `(index, valeur)`.
- `zip(a, b, ...)` permet d'itérer en parallèle sur plusieurs itérables.

### 5.1. `range()`

**Définition :**

`range()` génère une **suite de nombres** entiers.
Il est souvent utilisé pour **répéter une action un certain nombre de fois**, ou pour parcourir des **indices**.

**Syntaxe :**

```python
range(stop)
range(start, stop)
range(start, stop, step)
```

* `start` → valeur de départ (inclus)
* `stop` → valeur d'arrêt (exclus)
* `step` → pas entre chaque valeur (positif ou négatif)

**💡 Astuce :**

`range()` ne crée **pas une vraie liste** en mémoire :
c'est un **itérable paresseux** (il génère les nombres au fur et à mesure), ce qui le rend très efficace.

`````python
# De 0 à 4
for k in range(5):
    print(k)  # 0 1 2 3 4

print("\n———")

# De 2 à 9 avec un pas de 3
for k in range(2, 10, 3):
    print(k)  # 2 5 8

print("\n———")

# De 10 à 1 (à rebours)
for k in range(10, 0, -2):
    print(k)  # 10 8 6 4 2
`````

### 5.2. `enumerate()`



**Définition :**

`enumerate()` permet de **récupérer à la fois l'index et la valeur** lors d'une boucle sur une séquence.

**Syntaxe :**

```python
enumerate(itérable, start=0)
```

* `itérable` → la séquence à parcourir (liste, chaîne, etc.)
* `start` → valeur de départ de l'index (par défaut 0)

**💡 Pourquoi c'est mieux :**

Sans `enumerate`, on devrait gérer un compteur manuellement :

```python
# Moins lisible
i = 1
for nom in noms:
    print(i, nom)
    i += 1
```

`````python
noms = ["Ana", "Ben", "Chloé"]

# enumerate(noms) =>
# [
#   (0, "Ana"),
#   (1, "Ben"),
#   (2, "Cholé")
# ]

for i, nom in enumerate(noms):
    print(f"{i}. {nom}")
`````

### 5.3. `zip()`

**Définition :**

`zip()` permet de **parcourir plusieurs séquences en parallèle**.
Chaque itération renvoie un **tuple** contenant les éléments correspondants.

**Syntaxe :**

```python
zip(itérable1, itérable2, ...)
```

**Si les listes n’ont pas la même longueur :**

`zip()` s’arrête **au plus court des itérables** :

```python
produits = ["Livre", "Stylo", "Cahier", "Trousse"]
prix = [10, 2, 5]

for p, pr in zip(produits, prix):
    print(f"{p} → {pr}€")
```

➡️ Le dernier élément ("Trousse") n'est **pas traité**.


`````python
produits = ["Livre", "Stylo", "Cahier", "Trousse"]
prix = [10, 2, 5]
stocks = [12554,3545,32437,5467]

# zip(produits, prix, stocks) =>
# [
#   ("Livre", 10, 12554),
#   ("Stylo", 2, 3545),
#   ("Cahier", 5, 5467)
# ]

for produit, prix_unitaire, stock in zip(produits, prix, stocks):
    print(f"{produit} → {prix_unitaire}€")
`````

### Bonus : Combinaison

`````python
produits = ["Livre", "Stylo", "Cahier", "Stylo"]
prix = [10, 2, 5, 999]

for i, (p, pr) in enumerate(zip(produits, prix), start=1):
    print(f"{i}. {p} coûte {pr}€")
`````

### En résumé



| Fonction      | Sert à...                           | Exemple typique                | Sortie           |
| ------------- | ----------------------------------- | ------------------------------ | ---------------- |
| `range()`     | Générer une suite de nombres        | `for i in range(3)`            | `0, 1, 2`        |
| `enumerate()` | Avoir index + valeur                | `for i, v in enumerate(liste)` | `(0, valeur)`    |
| `zip()`       | Parcourir plusieurs listes ensemble | `for a, b in zip(l1, l2)`      | `(l1[0], l2[0])` |

## 6. La clause `else` sur les boucles

En Python, une boucle `for` ou `while` peut avoir un `else` qui s'exécute **si la boucle n'a pas été interrompue par `break`**.

`````python
# Recherche d'un élément; 'else' s'exécute si aucun 'break' n'a eu lieu
liste = [2, 4, 6, 8]
x = 1
for n in liste:
  if n == x:
    print("Trouvé !")
    break
else:
  print("Non trouvé.")
`````

## 7. Bonnes pratiques

1. **Éviter de modifier** une liste pendant qu'on l'itère → itérer sur une **copie** (`for x in list(lst): ...`).
2. Préférer `enumerate` plutôt que de gérer un index séparé.
3. Utiliser `zip` pour les itérations parallèles plutôt que des index manuels.
4. Utiliser des **conditions nommées** si la logique est complexe.
5. Préférer `for` à `while` quand on itère sur une collection/une progression connue.

`````python
## 1. Éviter de modifier une liste pendant qu'on l'itère → itérer sur une copie

nombres = [1, 2, 3, 4, 5]

# ❌ Mauvais : on modifie la liste pendant qu’on la parcourt
# for n in nombres:
#     if n % 2 == 0:
#         nombres.remove(n)  # provoque des comportements inattendus

# ✅ Bon : on parcourt une copie pour éviter les effets de bord
for n in list(nombres):  # list() crée une copie
    if n % 2 == 0:
        nombres.remove(n)

print("Liste finale :", nombres)  # [1, 3, 5]
`````

`````python
## 2. Préférer enumerate plutôt que de gérer un index séparé

fruits = ["pomme", "banane", "cerise"]

# ❌ Mauvais : on gère l’index à la main
i = 0
for fruit in fruits:
    print(i, fruit)
    i += 1

# ✅ Bon : enumerate fait le travail
for i, fruit in enumerate(fruits):
    print(i, fruit)
`````

`````python
## 3. Utiliser zip pour les itérations parallèles plutôt que des index manuels

produits = ["Livre", "Stylo", "Cahier"]
prix = [10, 2, 5]

# ❌ Mauvais : on manipule les indices
for i in range(len(produits)):
    print(produits[i], "→", prix[i], "€")

# ✅ Bon : zip assemble les listes
for p, pr in zip(produits, prix):
    print(p, "→", pr, "€")
`````

`````python
## 5. Préférer for à while quand on itère sur une collection/une progression connue

# ❌ while est inutilement complexe ici
nombres = [10, 20, 30, 40]
i = 0
while i < len(nombres):
    print("Élément :", nombres[i])
    i += 1

# ✅ for est plus clair et plus pythonique
for n in nombres:
    print("Élément :", n)
`````
