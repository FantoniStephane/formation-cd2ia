# Les environnements virtuels en Python

Quand on développe en Python, on installe souvent des bibliothèques externes avec `pip` : par exemple `requests`, `flask` ou `django`.

Le problème, c'est que si tous les projets utilisent les mêmes bibliothèques installées au même endroit sur l'ordinateur, on risque vite d'avoir des conflits :
- un projet a besoin de `Django 4`
- un autre a besoin de `Django 5`
- un troisième ne doit surtout pas être cassé par une mise à jour

Pour éviter cela, on utilise un **environnement virtuel**.

Un environnement virtuel est un **petit espace Python isolé**, propre à un projet.  
Il contient :
- son propre interpréteur Python
- ses propres paquets installés avec `pip`

Ainsi, chaque projet peut avoir ses dépendances sans perturber les autres.


### Pourquoi c'est utile ?
- éviter les conflits de versions
- garder un projet propre et reproductible
- pouvoir partager facilement les dépendances avec un autre développeur
- retrouver plus facilement un environnement de travail identique sur une autre machine

### Cycle classique

Dans un projet Python, on fait souvent toujours les mêmes étapes :
1. créer un environnement virtuel
2. l'activer
3. installer les bibliothèques nécessaires
4. travailler sur le projet
5. exporter les dépendances avec `pip freeze`
6. désactiver l'environnement

### Créer un environnement virtuel

La commande la plus classique est :

`python -m venv .venv`

Ici :
- `python -m venv` lance le module standard de Python prévu pour cela
- `.venv` est le nom du dossier qui contiendra l'environnement virtuel

On utilise souvent le nom `.venv` car il est clair, courant, et souvent bien reconnu par les IDE.

### Activer l'environnement

Une fois créé, il faut l'activer.

Sous **macOS / Linux** :
`source .venv/bin/activate`

Sous **Windows (PowerShell)** :
`.venv\Scripts\Activate.ps1`

Quand l'environnement est activé, le terminal affiche souvent son nom au début de la ligne, par exemple :

`(.venv)`

Cela signifie que les commandes `python` et `pip` pointent maintenant vers cet environnement, et non vers l'installation globale de la machine.


### Installer des dépendances

Une fois l'environnement activé, on peut installer les bibliothèques nécessaires au projet.

Exemples :
- `pip install requests`
- `pip install flask`
- `pip install django`

La bibliothèque sera installée **dans l'environnement virtuel courant**, pas globalement sur l'ordinateur.

### Voir les paquets installés

Pour afficher les dépendances installées dans l'environnement :

`pip list`


### Exporter les dépendances : pip freeze

Quand le projet fonctionne, on peut sauvegarder la liste exacte des paquets installés avec :

`pip freeze > requirements.txt`

Cela crée un fichier `requirements.txt` qui pourra être réutilisé plus tard.

Exemple de contenu :

```txt
Flask==3.1.0
requests==2.32.3
python-dotenv==1.0.1
```

#### Rappel des commandes :

```bash
# Exporter les dépendances dans un fichier
pip freeze > requirements.txt

# Afficher le contenu du fichier sous macOS / Linux
cat requirements.txt

# Afficher le contenu du fichier sous Windows PowerShell
Get-Content requirements.txt
```


### Réinstaller les dépendances d'un projet existant

Si un projet contient déjà un fichier `requirements.txt`, on peut recréer l'environnement puis installer tout d'un coup :

`pip install -r requirements.txt`

#### Rappel des commandes

```bash
# Recréer l'environnement
python -m venv .venv

# L'activer
source .venv/bin/activate

# Réinstaller toutes les dépendances du projet
pip install -r requirements.txt
```

### Désactiver l'environnement

Quand on a terminé :

`deactivate`

On revient alors au Python global de la machine.


### Bonnes pratiques

- créer **un environnement virtuel par projet**
- ne pas versionner le dossier `.venv` dans Git
- versionner le fichier `requirements.txt`
- activer l'environnement avant de lancer le projet ou d'installer un paquet


### Exemple concret

Imaginons deux projets différents :

- `projet_a` utilise `Flask`
- `projet_b` utilise `Django`

Sans environnement virtuel, toutes les installations iraient au même endroit, ce qui pourrait provoquer des conflits.

Avec un environnement virtuel par projet :
- `projet_a/.venv` contient uniquement ce dont `projet_a` a besoin
- `projet_b/.venv` contient uniquement ce dont `projet_b` a besoin

Les deux projets peuvent donc coexister proprement sur la même machine.


> **Attention :** `pip` installe les paquets dans l'environnement actuellement actif.  
> Si l'environnement virtuel n'est pas activé, `pip` risque d'installer les dépendances globalement sur la machine.


### En résumé

Un environnement virtuel permet d'avoir un projet Python **isolé, propre et reproductible**.  
C'est une habitude indispensable dès qu'on travaille avec `pip` et des bibliothèques externes.
