## 1. Votre première classe


Une **classe** permet de fabriquer des objets (ou des instances).

On peut la voir comme un **modèle** ou un **plan** qui décrit ce que les objets auront en commun.

Ici, `Car` est la **classe**.
Elle sert à définir le type d'objet que l'on veut créer.

`car_1` et `car_2` sont deux **objets différents** (ou instances différentes) créés à partir de cette même classe.

Ils ont donc le même modèle de départ, mais ce sont bien **deux éléments distincts** dans le programme, avec chacun leurs propres données.

```python
# On définit une classe appelée Car
class Car:
    pass # On laisse la classe vide pour l'instant.

# On crée deux instances de cette classe
car_1 = Car()
car_2 = Car()

# Est-ce que ce sont les mêmes objets ?
print("Mêmes objets ?", car_1 == car_2)

# Est-ce qu'ils ont été créés à partir de la même classe ?
print("Même classe ?", type(car_1) == type(car_2))
```

## 2. Définir une méthode dans une classe


Pour créer une **méthode**, on écrit une **fonction à l'intérieur de la classe** avec le mot-clé `def`.

Dans l'exemple ci-dessous, la méthode définit simplement une **action** que les objets de type `Car` pourront effectuer.

```python
# Reprenons notre classe Car
class Car:
    def start():
        print("Vroum... 🚗")

# On crée une instance
my_supercar = Car()

# On démarre la voiture
my_supercar.start() # Quand on écrit `my_supercar.start()`, Python comprend que la méthode `start` doit s'exécuter pour l'objet `my_supercar`.
```

### 💡 À retenir

> Une méthode est une fonction écrite dans une classe avec `def`.
> Elle décrit une action que les objets de cette classe pourront réaliser.

## 3. Ajouter un constructeur avec `__init__`


`__init__` est une méthode spéciale exécutée automatiquement au moment où l'objet est créé (ici, le nom est important. Il n'est pas arbitrairement choisi).

Elle sert en général à **donner ses premières valeurs à l'objet**, par exemple sa marque, sa couleur ou son prix.

`self` désigne l'objet en cours de création ou de manipulation.
C'est grâce à lui qu'on peut stocker des informations **dans cet objet précis**. Dans la définition des paramètres d'une méthode, il sera systématiquement le **premier** paramètre.

Ainsi, avec `self.marque` et `self.couleur`, on enregistre ces données **à l'intérieur de l'objet**.

Cela permet à chaque objet de conserver ses **propres valeurs**.

Par exemple, si on crée deux voitures, l'une peut être `"Peugeot"` et l'autre `"Renault"` : chaque objet garde ses informations séparément.

```python
# On utilise toujours notre classe Car
class Car:
  # On définit une méthode qui s'appelle __init__
  def __init__(self, brand, color):
    # On remplit les valeurs de la classe
    self.brand = brand
    self.color = color

# On instancie les objets en spécifiant les paramètre dans l'ordre
# où ils ont été définis.
car_1 = Car("Renault", "rouge")
car_2 = Car("Tesla", "noire")

# Testons les accès !
print(f"- Marque de la voiture 1 : {car_1.brand}")
print(f"- Couleur de la voiture 1 : {car_1.color}")
print(f"- Marque de la voiture 2 :{car_2.brand}")
```

## 4. Modifier l'état d'un objet


L'objet peut donc évoluer dans le temps.
Sa donnée `kilometers` initialisée dans le constructeur change après appel de la méthode `drive()`.


```python
# On utilise toujours notre classe Car
class Car:
  # On définit une méthode qui s'appelle __init__
  def __init__(self):
    # On remplit les valeurs de la classe
    self.kilometers = 0

  # On définit une méthode pour augmenter le compteur kilométrique
  def drive(self, kms: int):
    self.kilometers += kms

# On instancie les objets en spécifiant les paramètre dans l'ordre
# où ils ont été définis.
my_supercar = Car()

print(f"Compteur kilométrique avant mon trajet : {my_supercar.kilometers}")
my_supercar.drive(450)
print(f"Compteur kilométrique à la pause sur l'aire d'autoroute : {my_supercar.kilometers}")
my_supercar.drive(856)
print(f"Compteur kilométrique à la fin du trajet : {my_supercar.kilometers}")
```

## 5. Chaque objet a son propre état


Même si les objets viennent de la même classe, ils gardent chacun leurs propres valeurs.


```python
# On utilise toujours notre classe Car
class Car:
  # On définit une méthode qui s'appelle __init__
  def __init__(self, model):
    # On remplit les valeurs de la classe
    self.kilometers = 0
    self.model = model

  # On définit une méthode pour augmenter le compteur kilométrique
  def drive(self, kms: int):
    self.kilometers += kms

car_1 = Car("Tesla Model 3")
car_2 = Car("Peugeot 3008")

# Les deux voitures roulent...
car_1.drive(155)
car_1.drive(300)
car_2.drive(10)
car_1.drive(685)
car_2.drive(40)

print(f"La voiture '{car_1.model}' a roulé {car_1.kilometers}km")
print(f"La voiture '{car_2.model}' a roulé {car_2.kilometers}km")
```

## 6. Un autre exemple...


```python
# On définit une classe pour gérer un compte bancaire
class BankAccount:
  def __init__(self, owner, balance):
    self.owner = owner
    self.balance = balance

  def deposit(self, amount):
    self.balance = self.balance + amount

  def withdraw(self, amount):
    self.balance = self.balance - amount

  def display_balance(self):
    print("Owner:", self.owner)
    print("Balance:", self.balance, "euros")

account = BankAccount("Nicolas", 100)
account.deposit(1500)
account.withdraw(500)
account.display_balance()
```
