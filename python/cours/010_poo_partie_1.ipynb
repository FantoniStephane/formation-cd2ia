## 1. Un objet, c'est quoi ?

Un **objet**, c'est une **entité du programme** qui sert à représenter une chose, avec à la fois :

* des **données** sur cette chose
* des **actions** que l'on peut faire avec elle

### Analogie simple

On peut penser à un **compte bancaire**.

Il a des **informations** :

* un titulaire
* un solde

Et on peut faire des **actions** dessus :

* déposer de l'argent
* retirer de l'argent
* afficher le solde

En programmation, un objet fonctionne un peu pareil : c'est une chose qui contient des données et sur laquelle on peut agir.

### Exemple en Python

```python
my_list = [1, 2, 3]
```

Ici, `my_list` est un **objet**.

Pourquoi ?

Parce qu'elle contient des **informations** :

* `1`
* `2`
* `3`

et elle permet aussi des **actions** :

```python
my_list.append(4)
my_list.sort()
```

### 💡 Définition simple à retenir

> Un **objet** est une chose manipulée par le programme, qui contient des informations et permet certaines actions.



## 2. Les chaînes de caractères sont aussi des objets


````python
# On lui assigne une valeur
text = "Hello Python !"

# Méthodes à disposition
print(text.upper())
print(text.replace("Python", "Javascript"))
print(text.capitalize())
````

Ici, `text` est une variable qui contient **un objet** de type `str` (pour `string`, une chaîne de caractères).

Il possède donc :
- une **valeur**
- des **méthodes** comme `upper()`, `replace()`, `capitalize()`


## 3. Regarder les capacités d'un objet


````python
text = "Hello World !"

text_capabilities = dir(text)
for capability in text_capabilities:
  print(f"Une des fonctions de text est : {capability}")
````

`dir(...)` affiche les éléments disponibles sur un objet, notamment ses **méthodes**.

La documentation officielle Python présente aussi ces informations, par exemple pour [`str`](https://docs.python.org/3/library/stdtypes.html#text-sequence-type-str) et pour [`list`](https://docs.python.org/3/tutorial/datastructures.html).

Vous n'avez pas besoin de tout retenir dans cette liste.
Ici, le but est surtout de constater qu'un objet n'est pas seulement une valeur.

L'idée importante est la suivante :

**un objet a un type, des données, et des méthodes**.

* le **type** indique ce qu'il est
* les **données** correspondent à ce qu'il contient
* les **méthodes** correspondent à ce qu'il sait faire


## 4. Classe vs objet


Une **classe** est un **modèle**.
Elle décrit ce qu'une chose va contenir et ce qu'elle pourra faire.

Un **objet** est un **exemplaire concret** créé à partir de ce modèle. Vous entendrez aussi le terme **instance** : dans ce contexte, cela désigne la même chose.

### Analogie simple

* **Classe** = le plan d'une maison
* **Objet** = une vraie maison construite à partir de ce plan, qui a des habitants, une adresse...

### En programmation

Si on crée une classe `Car`, on décrit une voiture de manière générale :

* elle peut avoir une couleur
* elle peut avoir une marque
* elle peut démarrer

Ensuite, une voiture précise comme `my_own_car` est un **objet**.

### 💡 À retenir

> **La classe décrit.**
> **L'objet est l'élément concret utilisé dans le programme.**
