# Découvrir le fonctionnement d'un notebook

Bienvenue dans ce premier notebook.

Un notebook permet de réunir dans un même document :
- du texte explicatif
- du code exécutable
- les résultats du code
- parfois des graphiques, tableaux ou images

Dans ce notebook, nous allons voir comment cela fonctionne concrètement.

> ⚠️ Si vous avez atteint ce notebook en copiant le lien partagé, vous n'avez qu'un accès en **lecture seule**. Pour en récupérer une copie sur votre Google Drive, cliquez sur *Fichier*, puis *Enregistrer une copie dans Drive*

## 1. Un notebook mélange texte et code

Dans un notebook, certaines cellules servent à **expliquer** une notion.

D'autres servent à **écrire du code Python**.

Cela permet d'apprendre pas à pas :
1. on lit une explication
2. on exécute le code
3. on observe le résultat
4. on modifie et on recommence

```python
# Pour exécuter le code, cliquez sur le petit icône "Play" à gauche de cette cellule.
print("Bonjour et bienvenue dans Google Colab !")
```

### Ce qu'il faut remarquer

Quand on exécute une cellule de code :
- le code s'exécute
- le résultat apparaît juste en dessous
- on peut relancer la cellule autant de fois que nécessaire

Un notebook est donc un document **vivant**.

```python
nom = "Python"
annee = 1991

print("Langage :", nom)
print("Année de diffusion :", annee)
```

## 2. On peut tester et modifier facilement

L’intérêt d'un notebook, c'est qu'on peut changer une valeur puis relancer la cellule immédiatement.

Essayez par exemple de remplacer dans la cellule de code ci-dessus :

- `"Python"` par `"Colab"`
- `1991` par une autre année

Puis relancez la cellule.

## 3. L'ordre d'exécution est important

Dans un notebook, les cellules ne sont pas forcément exécutées de haut en bas automatiquement.

C’est vous qui choisissez quelles cellules lancer.

Cela signifie qu'une cellule peut dépendre d'une autre exécutée avant.

Exemple : si une variable n'a pas encore été créée, Python ne peut pas l'utiliser.

```python
message = "Je suis une variable créée dans cette cellule."
print(message)
```

```python
print("Contenu de la variable message :")
print(message) # Ici, "message" fait référence à la variable créée dans la cellule précédente.
```

### Expérience à faire

Essayez de :
1. redémarrer l'environnement d’exécution
2. lancer directement la cellule précédente
3. observer ce qui se passe

Vous verrez qu'un notebook garde en mémoire ce qui a déjà été exécuté... tant que la session est active.

## 4. Un notebook peut aussi afficher des listes et des boucles

Le notebook est pratique pour tester rapidement de petites instructions Python.

```python
prenoms = ["Lina", "Noah", "Sarah", "Adam"]

for prenom in prenoms:
    print("Bonjour", prenom)
```

## 5. On peut produire des graphiques

Un notebook ne sert pas seulement à afficher du texte.

Il peut aussi afficher :
- des graphiques
- des tableaux
- des images
- des résultats d’analyse

C'est l'une des raisons pour lesquelles ce format est très utilisé en data, en IA et en enseignement.

```python
import matplotlib.pyplot as plt

notes = [12, 15, 9, 17, 14]
eleves = ["A", "B", "C", "D", "E"]

plt.bar(eleves, notes)
plt.title("Exemple de graphique dans un notebook")
plt.xlabel("Élèves")
plt.ylabel("Notes")
plt.show()
```

## 6. On peut documenter le raisonnement

Le gros avantage du notebook, ce n'est pas seulement d'exécuter du code.

C'est aussi de :
- garder une trace du raisonnement
- expliquer chaque étape
- rendre le travail plus lisible
- mélanger théorie et pratique dans un même support

Autrement dit, un notebook n'est pas juste un fichier de code :
c'est un support de travail, d'apprentissage et de démonstration.

## 7. 🎯 Mini exercice

Modifiez le code ci-dessous pour :
1. changer le nom de la matière
2. changer la note
3. afficher un message différent si la note est supérieure ou égale à 10

```python
matiere = "mathématique"
note = 19

print("Matière :", matiere)
print("Note :", note)

if note >= 10:
    print("BRAVO")
else:
    print("Résultat : à retravailler")
```

## 8. Ce qu'il faut retenir

Un notebook permet de :
- lire des explications
- écrire du code
- exécuter cellule par cellule
- voir les résultats immédiatement
- documenter une démarche complète

C'est pour cela que nous l'utiliserons souvent dans ce module.

# Fin du notebook

Vous venez de voir les grands principes de fonctionnement d'un notebook :

- texte + code + résultat
- exécution progressive
- importance de l'ordre des cellules
- possibilité d’ajouter des visualisations

Le notebook est donc un excellent support pour apprendre en pratiquant.
