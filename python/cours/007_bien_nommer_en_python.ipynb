## 1. Pourquoi le nommage est important

> *"Any fool can write code that a computer can understand. Good programmers write code that humans can understand."*  
> — **Martin Fowler**

En tant que développeur, vous passez **beaucoup plus de temps à lire du code qu'à en écrire**. Un bon nommage transforme un code cryptique en une histoire lisible.

Un bon nom permet de :
- **Comprendre le code sans commentaires** : le nom se suffit à lui-même
- **Faciliter la maintenance** : vous (ou un collègue) retrouverez facilement la logique 6 mois plus tard
- **Réduire les bugs** : un nom clair révèle les incohérences logiques
- **Collaborer efficacement** : tout le monde parle le même langage dans le code

### Exemple concret : nommage obscur vs expressif

Comparez ces deux versions du **même programme** :

```python
# ❌ Nommage obscur
def f(l):
    s = 0
    for x in l:
        if x > 0:
            s += x
    return s / len(l)


# Test
n = [15, -1, 12, 0, 18]
print(f(n))
```

```python
# ✅ Nommage expressif
def calculate_positive_average(grades):
    total = 0
    for grade in grades:
        if grade > 0:
            total += grade
    return total / len(grades)

# Test
notes = [15, -1, 12, 0, 18]
print(calculate_positive_average(notes))
```

Le code de droite se lit presque comme une phrase : *"calculer la moyenne positive des notes"*. Aucun commentaire n'est nécessaire, les noms font tout le travail.

## 2. Les conventions PEP 8 en un coup d'œil

La [PEP 8](https://peps.python.org/pep-0008/#naming-conventions) est le guide de style officiel de Python.

| Élément | Convention | Exemple |
|---|---|---|
| Variable | `snake_case` | `user_age`, `total_price` |
| Constante | `UPPER_SNAKE_CASE` | `MAX_RETRIES`, `API_URL` |
| Fonction / Méthode | `snake_case` + verbe | `calculate_total()`, `send_email()` |
| Classe | `PascalCase` (nom) | `ShoppingCart`, `UserProfile` |
| Module | `snake_case` court | `data_utils.py`, `validators.py` |
| Package | minuscules, sans underscore | `mypackage`, `requests` |
| Attribut/méthode privé(e) | `_prefixe` | `_validate_input()`, `_cache` |
| Name mangling | `__double_prefixe` | `__internal_state` |
| Méthode "magique" | `__dunder__` | `__init__`, `__str__`, `__len__` |

### Noms à éviter ⚠️

- Les lettres seules `l` (L minuscule), `O` (O majuscule) et `I` (I majuscule), trop facilement confondues avec `1` et `0`
- Les noms à une seule lettre en dehors des boucles courtes (`i`, `j`) ou des conventions mathématiques (`x`, `y`, `n`)
- Les noms qui masquent des built-ins Python : `list`, `dict`, `str`, `type`, `id`, `input`, `sum`, `min`, `max`

```python
# Exemple illustrant toutes les conventions dans un même fichier

# Constantes (UPPER_SNAKE_CASE)
MAX_LOGIN_ATTEMPTS = 5
DEFAULT_TIMEOUT = 30

# Classe (PascalCase)
class UserAccount:
    # Méthode magique (__dunder__)
    def __init__(self, username, email):
        # Attributs d'instance (snake_case)
        self.username = username
        self.email = email
        # Attribut privé (_prefixe)
        self._login_count = 0

    # Méthode publique (snake_case + verbe)
    def reset_password(self, new_password):
        self._validate_password(new_password)
        # self._update_password(new_password)  # méthode non définie ici

    # Méthode privée (_prefixe)
    def _validate_password(self, password):
        is_long_enough = len(password) >= 8
        has_digit = any(char.isdigit() for char in password)
        return is_long_enough and has_digit

# Test
account = UserAccount("alice", "alice@example.com")
print(f"Utilisateur : {account.username}")
print(f"Mot de passe valide ? {account._validate_password('SecurePass1')}")
```

## 3. Comment trouver un bon nom : la méthode

### 3.1 Se poser les bonnes questions

Avant de nommer quoi que ce soit, posez-vous ces 3 questions :

1. **Qu'est-ce que cette variable contient / que fait cette fonction ?**  
   Décrivez-le en une phrase simple en français, puis traduisez en anglais.
2. **Si quelqu'un lit ce nom sans contexte, comprend-il son rôle ?**  
   Si non, le nom n'est pas assez explicite.
3. **Le nom est-il assez spécifique sans être trop long ?**  
   Il doit être précis sans devenir une phrase entière.

### 3.2 Variables : nommer ce que ça contient

Le nom d'une variable doit répondre à la question : *"Qu'est-ce qu'il y a dedans ?"*

```python
# ✅ Le nom décrit le contenu
total_price = 49.99
users = ["Alice", "Bob", "Charlie"]
user_list = ["Alice", "Bob", "Charlie"]
error_message = "Invalid email address"
remaining_attempts = 3

print(total_price, users, error_message, remaining_attempts)
```

### 3.3 Booléens : utiliser un préfixe

Un booléen représente une condition vraie ou fausse. Son nom doit se lire comme une **affirmation** :

```python
# ✅ Préfixes courants : is_, has_, can_, should_
is_active = True
has_permission = False

user_role = "admin"
can_edit = user_role == "admin"

attempt_count = 2
MAX_RETRIES = 5
should_retry = attempt_count < MAX_RETRIES

# Lorsque vous lisez `if is_active:`, ça se lit naturellement : "si est actif"
if is_active:
    print("L'utilisateur est actif")

print(f"Peut éditer : {can_edit}")
print(f"Doit réessayer : {should_retry}")
```

### 3.4 Fonctions et méthodes : verbe + complément

Une fonction **fait quelque chose**. Son nom doit commencer par un **verbe d'action** suivi d'un complément qui précise sur quoi elle agit.

```python
def get_user_by_id(user_id):
    # Récupère un utilisateur par son identifiant
    return {"id": user_id, "name": "Alice"}

def calculate_shipping_cost(weight, destination):
    # Calcule le coût de livraison
    return weight * 0.5 + (10 if destination == "international" else 3)

def validate_phone_number(phone):
    # Valide un numéro de téléphone (exemple simplifié)
    return phone.startswith("+") and len(phone) >= 10

def is_expired(expiry_date):
    # Retourne un booléen → nom commence par is_
    from datetime import date
    return expiry_date < date.today()

# Tests
print(get_user_by_id(42))
print(calculate_shipping_cost(2.5, "local"))
print(validate_phone_number("+33612345678"))
```

### Verbes courants à garder en tête

| Verbe | Usage |
|---|---|
| `get` / `fetch` | Récupérer une donnée |
| `set` / `update` | Modifier une valeur |
| `create` / `build` | Créer un objet |
| `delete` / `remove` | Supprimer |
| `find` / `search` | Chercher |
| `filter` / `sort` | Filtrer / trier |
| `validate` / `check` | Vérifier |
| `parse` / `extract` | Analyser / extraire |
| `format` / `convert` | Formater / convertir |
| `compute` / `calculate` | Calculer |
| `send` / `notify` | Envoyer / notifier |
| `handle` / `process` | Traiter |

### 3.5 Classes : un nom (PascalCase) qui décrit l'entité

Une classe représente un **concept** ou une **entité**. Son nom est un **nom commun au singulier** :

```python
class Invoice:              # Une facture
    pass

class UserProfile:          # Un profil utilisateur
    pass

class DatabaseConnection:   # Une connexion à la base de données
    pass

class EmailValidator:       # Un validateur d'email
    def validate(self, email):
        return "@" in email and "." in email

# Test
validator = EmailValidator()
print(validator.validate("alice@example.com"))  # True
print(validator.validate("pas-un-email"))        # False
```

### 3.6 Longueur du nom : trouver l'équilibre

| Qualité | Exemple | Problème |
|---|---|---|
| Trop court | `d` | Incompréhensible hors contexte |
| Trop long | `the_total_number_of_days_since_account_creation` | Verbeux, pénible à taper et à lire |
| Juste bien | `days_since_creation` | Clair, concis, suffisamment descriptif |

**Règle d'or** : plus la portée d'une variable est large, plus son nom doit être descriptif.

```python
# ✅ Portée courte (boucle) → nom court OK
for i in range(10):
    print(i, end=" ")
print()

values = [1, 2, 3, 4, 5]
total = sum(x ** 2 for x in values)
print(f"Somme des carrés : {total}")
```

```python
# ✅ Portée large (module) → nom descriptif
MAX_LOGIN_ATTEMPTS = 5
DATABASE_CONNECTION_TIMEOUT = 30
DEFAULT_CURRENCY_CODE = "EUR"

print(MAX_LOGIN_ATTEMPTS, DATABASE_CONNECTION_TIMEOUT, DEFAULT_CURRENCY_CODE)
```

## 4. Exemples concrets : Avant / Après

### Exemple 1 : Variables

```python
# ❌ Avant
n = "Alice"
a = 25
l = ["Python", "JavaScript"]
flag = True
```

```python
# ✅ Après
username = "Alice"
user_age = 25
known_languages = ["Python", "JavaScript"]
is_verified = True

# Chaque variable dit maintenant ce qu'elle contient.
# `flag` est devenu `is_verified` : on sait immédiatement ce que ce booléen représente.
print(f"{username}, {user_age} ans, vérifié: {is_verified}, langages: {known_languages}")
```

### Exemple 2 : Fonctions

```python
# ❌ Avant — `process(data)` ne dit rien
def process(data):
    res = []
    for d in data:
        if d["status"] == "active":
            res.append(d)
    return res

# Test
all_us = [
    {"name": "Alice", "status": "active"},
    {"name": "Bob",   "status": "inactive"},
    {"name": "Carol", "status": "active"},
]
act = process(all_us)
print(act)
```

```python
# ✅ Après — le nom dit exactement ce que fait la fonction et sur quoi elle travaille
def filter_active_users(users):
    active_users = []
    for user in users:
        if user["status"] == "active":
            active_users.append(user)
    return active_users

# Test
all_users = [
    {"name": "Alice", "status": "active"},
    {"name": "Bob",   "status": "inactive"},
    {"name": "Carol", "status": "active"},
]
actifs = filter_active_users(all_users)
print(actifs)
```

### Exemple 3 : Classe

```python
# ❌ Avant — `Mgr` est une abréviation obscure, `do()` ne dit rien
class Mgr:
    def __init__(self, d):
        self.d = d
        self.c = 0

    def do(self, item):
        self.d.append(item)
        self.c += 1

# Test
invt = Mgr(["Clavier", "Souris"])
invt.do("Écran")
print(f"Produits : {invt.do}")
print(f"Nombre ajouté : {invt.c}")
```

```python
# ✅ Après — classe et méthode clairement nommées
class InventoryManager:
    def __init__(self, products):
        self.products = products
        self.product_count = 0

    def add_product(self, product):
        self.products.append(product)
        self.product_count += 1

# Test
inventory = InventoryManager(["Clavier", "Souris"])
inventory.add_product("Écran")
print(f"Produits : {inventory.products}")
print(f"Nombre ajouté : {inventory.product_count}")
```

## 5. Les erreurs fréquentes ❌

| Erreur | Mauvais exemple | Bonne pratique |
|---|---|---|
| Nommer en français | `calculer_prix_total` | `calculate_total_price` |
| Abréviations obscures | `calc_ttl_amt` | `calculate_total_amount` |
| Noms génériques | `data`, `result`, `tmp` | `user_records`, `filtered_orders` |
| Mélanger les conventions | `getUserName` (camelCase) | `get_user_name` (snake_case) |
| Noms trompeurs | `users` qui contient 1 user | `user` au singulier |
| Masquer des built-ins | `list = [1, 2, 3]` | `user_list = [1, 2, 3]` |

```python
# ⚠️ Démonstration du shadowing de built-in (à ne PAS faire)

# Mauvais : écrase la fonction built-in list()
# list = [1, 2, 3]  # ← décommentez pour voir le problème
# list("hello")      # ← TypeError : 'list' object is not callable

# ✅ Bon : nom spécifique
user_list = [1, 2, 3]
print(list("hello"))  # ['h', 'e', 'l', 'l', 'o'] — built-in intact
print(user_list)
```

## 6. Récapitulatif : Checklist ✅

Avant de valider un nom, vérifiez qu'il coche toutes ces cases :

- ✅ Le nom **décrit clairement** le contenu (variable) ou l'action (fonction)
- ✅ Il suit la **convention PEP 8** appropriée (`snake_case`, `PascalCase`, `UPPER_SNAKE_CASE`…)
- ✅ Il est écrit en **anglais**
- ✅ Il n'est **ni trop court** (`x`) **ni trop long** (`the_total_number_of…`)
- ✅ Un booléen commence par `is_`, `has_`, `can_` ou `should_`
- ✅ Une fonction commence par un **verbe d'action**
- ✅ Une classe est un **nom en PascalCase** au singulier
- ✅ Il ne **masque pas** un built-in Python (`list`, `dict`, `id`…)
- ✅ Il ne contient pas d'**abréviation obscure**

## 7. Exercice récapitulatif

Réécrivez le code ci-dessous en appliquant toutes les bonnes pratiques de nommage vues dans ce guide :

```python
# ❌ Code à améliorer
MX = 100

class Mgr2:
    def __init__(self, d):
        self.d = d
        self.flag = False

    def proc(self, x):
        res = []
        for i in x:
            if i > 0 and len(res) < MX:
                res.append(i)
        self.flag = len(res) > 0
        return res

# Test
filt = Mgr2([])
result = filt.proc([-3, 0, 5, 12, -1, 7])
print(f"Résultat : {result}")
print(f"A des résultats : {filt.flag}")
```

```python
# ✅ Votre version améliorée ici :

MAX_RESULTS = 100

class PositiveNumberFilter:
    def __init__(self, numbers):
        self.numbers = numbers
        self.has_results = False

    def filter_positive_numbers(self, values):
        positive_numbers = []
        for value in values:
            if value > 0 and len(positive_numbers) < MAX_RESULTS:
                positive_numbers.append(value)
        self.has_results = len(positive_numbers) > 0
        return positive_numbers

# Test
filterer = PositiveNumberFilter([])
result = filterer.filter_positive_numbers([-3, 0, 5, 12, -1, 7])
print(f"Résultat : {result}")
print(f"A des résultats : {filterer.has_results}")
```
