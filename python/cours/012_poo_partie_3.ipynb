## 1. Encapsulation


L'**encapsulation** consiste à **regrouper dans un même objet ses données et les méthodes qui agissent sur elles**, tout en **contrôlant la façon dont on peut modifier ces données**.

Autrement dit, on évite de laisser n'importe quelle partie du programme changer les valeurs directement.

On préfère passer par des méthodes prévues pour cela.

### Idée simple

Au lieu de faire n'importe quoi avec une donnée sensible, l'objet **protège son état** et impose une façon correcte de l'utiliser.

### Exemple

Pour un compte bancaire, on évite par exemple de modifier le solde n'importe comment.

On préfère utiliser des méthodes comme :

* `deposit()`
* `withdraw()`

Ainsi, l'objet garde le contrôle sur ses propres données.


```python
# On définit une classe pour représenter un compte bancaire
class BankAccount:
  def __init__(self, owner, balance):
    # owner : nom du titulaire du compte
    self.owner = owner

    # _balance : solde du compte
    # Le "_" au début de _balance indique que cet attribut
    # ne devrait pas être modifié directement depuis l'extérieur.
    # C'est une convention Python pour signaler :
    # "cet attribut est interne à l'objet".
    self._balance = balance

  def get_balance(self):
    # Retourne le solde actuel du compte
    return self._balance

  def deposit(self, amount):
    # Dépose de l'argent sur le compte si le montant est positif
    if amount > 0:
      self._balance = self._balance + amount

  def withdraw(self, amount):
    # Retire de l'argent si le montant est positif
    # et si le compte contient assez d'argent
    if amount > 0 and amount <= self._balance:
      self._balance = self._balance - amount


# On crée un objet BankAccount
account = BankAccount("Steve Jobs", 100)

# On affiche le solde initial
print(account.get_balance())

# On effectue un dépôt puis un retrait
account.deposit(50)
account.withdraw(20)

# On affiche le nouveau solde
print(account.get_balance())
```

Ici, on évite de manipuler le solde directement.
On passe par des méthodes prévues pour cela.


## 2. Héritage


L'**héritage** consiste à **créer une nouvelle classe à partir d'une classe existante**.

La nouvelle classe récupère automatiquement les **attributs** et les **méthodes** de la classe de départ, puis peut :

* **ajouter** de nouveaux éléments
* **modifier** certains comportements

### Idée simple

On part d'une classe générale, puis on crée une classe plus spécialisée.

Par exemple :

* `Animal` peut représenter tous les animaux
* `Dog` et `Cat` peuvent hériter de `Animal`

Ainsi, `Dog` et `Cat` récupèrent ce qui existe déjà dans `Animal`, sans avoir à tout réécrire.

### Exemple


```python
# Classe parent : elle contient un comportement commun
class Animal:
    def __init__(self, name):
        # Chaque animal possède un nom
        self.name = name

    def introduce(self):
        # Méthode commune à tous les animaux
        print("My name is", self.name)

    def sleep(self):
        # Autre méthode commune
        print(f"{self.name} is sleeping")


# Classe enfant : Dog hérite de Animal
class Dog(Animal):
    def bark(self):
        # Méthode spécifique au chien
        print("Woof")


# Classe enfant : Cat hérite de Animal
class Cat(Animal):
    def meow(self):
        # Méthode spécifique au chat
        print("Meow")


# On crée un objet Dog
dog = Dog("Rex")

# On crée un objet Cat
cat = Cat("Minou")

# Dog peut utiliser la méthode héritée de Animal
dog.introduce()
dog.sleep()

# Dog peut aussi utiliser sa méthode spécifique
dog.bark()

# Cat peut utiliser la méthode héritée de Animal
cat.introduce()
cat.sleep()

# Cat peut aussi utiliser sa méthode spécifique
cat.meow()
```

Ici :

* `Animal` est la **classe parent**
* `Dog` et `Cat` sont les **classes enfant**
* `Dog` et `Cat` héritent des méthodes de `Animal`
* Chaque classe enfant peut ajouter **ses propres méthodes**

## 3. Abstraction


L'**abstraction** consiste à définir un **modèle général** pour plusieurs classes, sans donner tous les détails de fonctionnement dès le départ. En Python, cela passe souvent par une **classe abstraite**, qui sert de base à d'autres classes.

Une classe abstraite **ne peut pas être instanciée directement**. Son rôle est surtout d'**imposer une structure commune** : les classes enfants devront compléter les méthodes prévues par cette classe de base.

En Python, on utilise pour cela le module `abc`, la classe `ABC` et le décorateur `@abstractmethod`.

```python
# On importe les outils nécessaires pour créer une classe abstraite
from abc import ABC, abstractmethod

# Classe abstraite : elle sert de modèle général
class Animal(ABC):
    def __init__(self, name):
        # Chaque animal possède un nom
        self.name = name

    @abstractmethod
    def make_sound(self):
        # Méthode abstraite :
        # on dit que tous les animaux doivent savoir produire un son,
        # mais on ne précise pas ici lequel
        pass

    def introduce(self):
        # Méthode classique, commune à tous les animaux
        print("My name is", self.name)


# Classe enfant : Dog hérite de Animal
class Dog(Animal):
    def make_sound(self):
        # Dog est obligé d'implémenter make_sound()
        print("Woof")


# Classe enfant : Cat hérite de Animal
class Cat(Animal):
    def make_sound(self):
        # Cat implémente aussi make_sound()
        print("Meow")


# On crée des objets à partir des classes concrètes
dog = Dog("Rex")
cat = Cat("Minou")

# Méthode héritée de la classe abstraite
dog.introduce()
cat.introduce()

# Méthode imposée par la classe abstraite
dog.make_sound()
cat.make_sound()


# Ceci ne fonctionnerait pas :
# animal = Animal("Test")
# Erreur : on ne peut pas créer directement un objet
# à partir d'une classe abstraite
```

### Ce qu'il faut retenir dans cet exemple

* `Animal` est une **classe abstraite**
* `make_sound()` est une **méthode abstraite**
* `Dog` et `Cat` doivent **obligatoirement** définir `make_sound()`
* on peut créer un objet `Dog` ou `Cat`, mais pas un objet `Animal` directement.


## 4. Polymorphisme


Le **polymorphisme** consiste à utiliser une **même méthode** sur des objets différents, tout en obtenant un **comportement adapté à chaque objet**.

Autrement dit, plusieurs objets peuvent répondre au **même message**, mais chacun à sa manière.

### Idée simple

Si plusieurs classes possèdent une méthode `make_sound()`, on peut appeler cette méthode sur différents objets sans avoir à savoir à l'avance s'il s'agit d'un chien, d'un chat ou d'un oiseau.

Chaque objet exécutera alors **sa propre version** de cette méthode.

```python
# Classe parent
class Animal:
    def __init__(self, name):
        # Chaque animal possède un nom
        self.name = name

    def make_sound(self):
        # Méthode prévue dans la classe parent
        return("Unknown sound")


# Classe enfant Dog
class Dog(Animal):
    def make_sound(self):
        # Le chien redéfinit la méthode à sa façon
        return("Woof")


# Classe enfant Cat
class Cat(Animal):
    def make_sound(self):
        # Le chat redéfinit aussi la même méthode
        return("Meow")


# Classe enfant Bird
class Bird(Animal):
    def make_sound(self):
        # L'oiseau a lui aussi sa propre version
        return ("Tweet")


# On crée plusieurs objets différents
dog = Dog("Rex")
cat = Cat("Minou")
bird = Bird("Rio")

# On les place dans une liste
animals = [dog, cat, bird]

# On parcourt la liste
for animal in animals:
    print(f"{animal.name} : {animal.make_sound()}")
```

### Ce qu'il faut comprendre dans cet exemple

* `Dog`, `Cat` et `Bird` ont tous une méthode `make_sound()`
* on appelle toujours **la même méthode**
* pourtant, le résultat change selon l'objet utilisé

C'est le **polymorphisme** :

* **même nom de méthode**
* **comportement différent selon l'objet**
