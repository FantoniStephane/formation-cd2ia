
# Découvrir et utiliser les modules Python

Dans ce notebook, vous allez apprendre à **comprendre**, **importer** et **utiliser** des modules en Python.  
L'objectif est de vous rendre autonome pour structurer votre code et réutiliser facilement des outils existants.


## 1. Comprendre ce qu'est un module

Un **module** est un fichier Python (`.py`) qui contient du code réutilisable : des **fonctions**, des **variables**, ou des **classes**.  
C'est un peu comme une **boîte à outils** : au lieu de tout réécrire à chaque fois, vous pouvez importer les outils déjà prêts.

### Pourquoi utiliser des modules ?

- Pour **éviter de répéter du code**.  
- Pour **mieux organiser** votre programme.  
- Pour **partager** facilement des fonctionnalités entre plusieurs fichiers.

### Où Python les trouve-t-il ?

Python cherche les modules à plusieurs endroits :
1. Dans les **modules intégrés** au langage (ex : `math`, `datetime`, `json`).
2. Dans les **fichiers de votre projet** (vos propres modules).
3. Dans les **modules installés** avec `pip` (modules "tiers").

Vous n'avez donc pas besoin de tout écrire : Python fournit déjà une grande bibliothèque prête à l’emploi.



## 2. Les modules natifs de Python

Les modules **natifs** sont inclus automatiquement avec Python.  
Vous pouvez les utiliser sans rien installer.

### Exemple 1 : Le module `math`


````python
import math

print("Racine carrée de 16 :", math.sqrt(16))
print("Cosinus de 0 :", math.cos(0))
print("La valeur de pi :", math.pi)
````


💡 Le module `math` contient de nombreuses fonctions mathématiques : racines, puissances, sinus, etc.  
C’est un bon exemple d’outil qu’on n’a pas besoin de recréer soi-même.


### Exemple 2 : Le module `datetime`

Ce module permet de manipuler les **dates et heures**.

````python
from datetime import datetime, timedelta

aujourdhui = datetime.now()
print("Date et heure actuelle :", aujourdhui)

dans_une_semaine = aujourdhui + timedelta(days=7)
print("Dans une semaine :", dans_une_semaine.strftime("%d/%B/%Y à %H:%M"))
````


### Exemple 3 : Le module `json`

Le format **JSON** (JavaScript Object Notation) est très utilisé pour échanger des données.  
Python permet de facilement convertir entre **objets Python** et **texte JSON**.


````python
import json

personne = {"nom": "Dupont", "age": 30, "ville": "Paris", "métier": None}

texte_json = json.dumps(personne) #dumps : dump string : dump : conversion pyhton->json; s: string : chaine de caractere
print("Texte JSON :", texte_json, type(texte_json))

donnees = json.loads(texte_json)
print("Nom extrait :", donnees["nom"])
````


### Exemple 4 : Le module `csv`

Les fichiers **CSV** (valeurs séparées par des virgules) sont souvent utilisés dans les tableurs (Excel, LibreOffice...).  
Le module `csv` permet d’écrire et lire ces fichiers.


````python
import csv

with open("eleves.csv", "w", newline="") as fichier:
    writer = csv.writer(fichier)
    writer.writerow(["Nom", "Âge", "Ville"])
    writer.writerow(["Alice", 22, "Metz"])
    writer.writerow(["Bob", 25, "Nancy"])

with open("eleves.csv", "r") as fichier:
    reader = csv.reader(fichier)
    for ligne in reader:
        print(ligne)

# Note : vous pouvez également utiliser l'onglet "Fichiers" de Google Colab pour ouvrir directement le fichier généré.
````


## 3. Créer et importer un module personnalisé

Vous pouvez créer vos **propres modules** pour organiser votre code.

Par exemple, si vous écrivez souvent des fonctions de conversion ou de calcul, vous pouvez les placer dans un fichier séparé.

### Exemple

1. Créez un fichier appelé `mon_module.py` dans le dossier de votre projet.
2. Écrivez-y ce code :
```python
def dire_bonjour(nom):
    return f"Bonjour {nom} !"

def addition(a, b):
    return a + b
```
3. Enregistrez le fichier, puis exécutez le code ci-dessous :


````python
import mon_module # Note : Colab vous précisera que ce module n'existe pas. C'est une erreur normale.

print(mon_module.dire_bonjour("Alice"))
print("Résultat de l'addition :", mon_module.addition(3, 4))
````


💡 **Astuce :**
- Le nom du fichier devient le nom du module (`mon_module.py` → `mon_module`).
- Le module doit être dans le **même dossier** ou dans un dossier accessible à Python.

## 5. Gérer les erreurs et adopter de bonnes pratiques

### Erreurs fréquentes

| Erreur | Cause probable | Solution |
|--------|----------------|-----------|
| `ModuleNotFoundError` | Le module n’est pas installé ou mal orthographié | Vérifiez le nom ou installez-le avec `pip install ...` |
| `ImportError` | Le module existe mais une fonction n’est pas trouvée | Vérifiez la syntaxe de l’import (`from module import fonction`) |
| `SyntaxError` | Import mal placé ou mauvaise indentation | Placez les `import` en haut du fichier |

### Bonnes pratiques

- Placez **tous vos imports au début** du fichier.  
- Séparez-les en trois blocs :
  1. Modules natifs (`import os`, `import json`)
  2. Modules tiers (`import requests`)
  3. Modules internes (`import mon_module`)
- Utilisez `as` pour raccourcir les noms longs :  
  `import numpy as np`


## 6. Bien choisir un module tiers

Avant d'installer un module tiers, il est important de **vérifier sa fiabilité**.

Tous les modules disponibles sur Internet ne se valent pas. Certains peuvent être **obsolètes**, **mal maintenus**, ou même **présenter des failles de sécurité**.

Voici quelques critères simples pour faire un bon choix :

1. **Le module est toujours maintenu.**
   Vérifiez la date de la **dernière mise à jour** sur [PyPI.org](https://pypi.org) ou sur GitHub.
   Un projet régulièrement mis à jour inspire confiance : cela signifie que les développeurs corrigent les bugs et s'adaptent aux nouvelles versions de Python.

2. **Le module a une communauté active.**
   Consultez le nombre d'**étoiles (stars)** sur GitHub, le volume de téléchargements ou les questions posées sur Stack Overflow.
   Plus un projet est utilisé, plus il est fiable et documenté.

3. **La documentation est claire et complète.**
   Une bonne documentation explique les principales fonctions, donne des exemples et détaille l'installation.
   Si la documentation est inexistante ou confuse, c'est souvent un mauvais signe.

4. **Le module n'est pas obsolète.**
   Si le site ou la documentation affiche un avertissement comme *“deprecated”* ou *“no longer maintained”*, il vaut mieux chercher une alternative plus récente.

5. **Le module est stable.**
   Un module peut être publié alors qu'il est encore en développement. Il risque par conséquent d'être **instable** (de nombreux changements majeurs - _breaking changes_ - fréquents). Pour vérifier la stabilité, consultez le journal des versions (_Changelog_) ou regardez le numéro de version (généralement, un numéro de version 0.X.Y indique une version alpha ou beta, instable).

6. **Pas de failles de sécurité connues.**
   Pour les modules utilisés en production, consultez les rapports de sécurité sur [https://pypi.org](https://pypi.org) ou sur le dépôt GitHub du projet.


💡 **En résumé :** avant d'installer un module, prenez 2 minutes pour vérifier qu'il est **récent, populaire, bien documenté et maintenu**.
Cela vous évitera bien des problèmes à long terme.



## 7. Lire une documentation

Jusqu'ici, vous avez appris à importer, utiliser et créer des modules.
Mais pour aller plus loin, vous devez savoir **lire une documentation officielle**.
C'est l'une des compétences les plus importantes d'un développeur.


### Pourquoi lire une documentation ?

La documentation est comme le **mode d'emploi** d'un module :

* Elle explique **à quoi il sert**  (ce qu'il fait et ce qu'il ne fait pas)
* Elle décrit **chaque fonction disponible**
* Elle indique **comment l'utiliser** correctement, avec des **exemples de code**.

Lire une documentation ne signifie pas tout comprendre immédiatement : il s'agit surtout de **savoir où chercher**.

⚠️ A noter que la plupart des documentations sont en anglais.
