## 3. Syntaxe de base : indentation et blocs

### 3.1 L'importance de l'indentation

En Python, on n'utilise pas d'accolades `{}` pour définir les blocs (comme en C, Java, JavaScript...).

À la place, **l'indentation** (les espaces au début de la ligne) indique à Python quelles lignes appartiennent au même bloc.

Exemple avec une condition :

````python
age = 18

if age == 18:
    # Cette ligne est indentée : elle fait partie du bloc "if"
    print("Tu es majeur.")

print("Fin du programme.")  # Cette ligne n'est plus dans le bloc "if"
````

### 3.2 Quelques règles simples

- On utilise généralement **4 espaces** pour indenter un bloc.
- Après `if`, `for`, `while`, `def`, etc. on met un **`:`** puis on va à la ligne et on **indente**.
- Toutes les lignes d'un même bloc doivent avoir **la même indentation**.

Exemple avec une boucle :

````python
for i in range(5):
    print("Tour", i)

print("La boucle est terminée.")
````

## 4. Programmes séquentiels : les instructions sont exécutées dans l'ordre

Un programme Python est lu **de haut en bas**, **ligne par ligne**.

Les instructions sont exécutées dans l'ordre où elles apparaissent (sauf cas particuliers comme les fonctions appelées plus tard).

Regardez l'exemple suivant et essayez de deviner ce qui s'affiche avant d'exécuter la cellule.

````python
print("Début du programme")
x = 10
y = 5
somme = x + y
print("La somme vaut", somme)
print("Fin du programme")
````

> **Illustration :**
> Un programme, c'est un peu comme une recette de cuisine : on suit les instructions **dans l'ordre**, étape par étape.

## 5. Les pièges classiques à éviter au début

Voici quelques erreurs très fréquentes quand on commence :

1. **Oublier les guillemets autour d'un texte**
   ```python
   print(Bonjour)  # Erreur
   print("Bonjour")  # Correct
   ```

2. **Confondre `=` et `==`**
   - `=` sert à **affecter** une valeur à une variable : `x = 5`
   - `==` sert à **comparer** deux valeurs : `x == 5`

3. **Oublier les deux-points `:` après `if`, `for`, etc.**
   ```python
   if x > 0  # Erreur : il manque ':'
       print("positif")
   ```

4. **Mauvaise indentation** (trop ou pas assez d'espaces)

5. **Accent et caractères spéciaux** dans le code
   - Dans les versions récentes de Python, les accents sont bien gérés.
   - Mais évitez d'utiliser des accents dans les **noms de variables**.

6. **Mélanger nombres et textes sans les convertir**
   ```python
   age = 18
   # print("Tu as " + age + " ans")  # Erreur : on concatène un texte et un nombre
   print("Tu as", age, "ans")        # Correct
   ```

## 6. `print()` et `input()`

Nous avons déjà vu `print()` qui sert à **afficher** quelque chose.

La fonction **`input()`** permet de **demander une saisie à l'utilisateur**.

### 6.1 Exemple simple avec `input()`

Exécutes le code ci-dessous et tape votre prénom dans la zone qui apparaît.

````python
prenom = input("Comment t'appelles-tu ? ")
print("Bonjour", prenom, "!")
````

### 6.2 Attention : `input()` renvoie toujours un **texte**

Même si l'utilisateur tape un nombre, Python le considère comme un **string** (texte).

Si on veut faire des calculs avec, il faut le convertir en entier avec `int()` ou en nombre décimal avec `float()`.

````python
texte_age = input("Quel âge as-tu ? ")

# Conversion du texte en entier
age = int(texte_age)

print("Dans 10 ans, tu auras", age + 10, "ans.")
````

> Si vous entrez quelque chose qui n'est pas un nombre (par exemple `bonjour`), Python affichera une erreur lors de la conversion avec `int()`.

## 7. Exemple de programme très simple

Nous allons écrire un petit programme qui :

1. Demande le prénom de l'utilisateur
2. Demande son âge
3. Calcule l'année où il aura 100 ans (approximation)
4. Affiche un message récapitulatif

Essayez d'abord de lire le code, puis exécute la cellule.

````python
prenom = input("Quel est ton prénom ? ")
age_texte = input("Quel âge as-tu ? ")

# On convertit l'âge en entier
age = int(age_texte)

# On fait une approximation grossière : on suppose qu'on est en 2026
annee_actuelle = 2026
annee_cent_ans = annee_actuelle + (100 - age)

print("Bonjour", prenom, "!")
print("Tu as", age, "ans.")
print("Tu auras 100 ans en", annee_cent_ans, ".")
````
