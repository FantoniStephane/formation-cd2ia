---

## Valeurs possibles

| Paramètre | Valeurs |
|-----------|--------|
| Météo | soleil, nuageux, vent, pluie |
| Zone | entrepot, parking, toiture, zone_securisee |
| Priorité | basse, normale, haute, critique |
| Batterie | 0 à 100 % |

---

## Règles de décollage

La mission est **refusée** si :
- Batterie < 20 %
- Météo = pluie
- Zone = zone_securisee sans autorisation spéciale

---

## Calcul du score de risque

| Condition | Points |
|-----------|--------|
| Météo = vent | +2 |
| Météo = nuageux | +1 |
| Zone = toiture | +2 |
| Zone = zone_securisee | +3 |
| Batterie entre 20 et 35 % | +2 |
| Priorité = critique | +2 |
| Priorité = haute | +1 |

**Interprétation :** 0-2 → faible · 3-5 → modéré · 6+ → élevé
