## Contexte

Développer un petit jeu en Python qui se joue dans le terminal.

Le principe est simple : le programme choisit un nombre secret, et l'utilisateur doit essayer de le deviner. À chaque proposition, le programme indique si le nombre recherché est plus grand, plus petit, ou si la bonne réponse a été trouvée.

---

## Objectif

Créer un programme Python capable de :
1. Stocker un nombre secret dans une variable
2. Demander à l'utilisateur de proposer un nombre
3. Répéter la question tant que le bon nombre n'a pas été trouvé
4. Comparer chaque proposition au nombre secret
5. Indiquer si le nombre secret est plus grand ou plus petit
6. Afficher un message de victoire si le bon nombre est trouvé

---

## Contraintes techniques

Ton programme doit obligatoirement contenir :
- Une variable pour stocker le nombre secret
- Un `input()` pour récupérer les propositions
- Un `int()` pour convertir la saisie
- Au moins une boucle `while`
- Au moins un `if`, un `elif`, un `else`
