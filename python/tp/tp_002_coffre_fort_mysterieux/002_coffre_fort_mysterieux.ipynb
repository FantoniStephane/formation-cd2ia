## Contexte

Développer un petit programme Python qui simule un système de sécurité permettant d'ouvrir ou non un coffre-fort.

---

## Objectif

Créer un programme Python capable de :
1. Demander le prénom de l'utilisateur
2. Demander un code secret
3. Demander l'âge de l'utilisateur
4. Vérifier si l'utilisateur est autorisé à ouvrir le coffre
5. Afficher un message personnalisé selon le résultat

---

## Règles

| Cas | Condition | Résultat |
|-----|-----------|----------|
| 1 | Code correct ET âge ≥ 18 | Accès autorisé |
| 2 | Code incorrect | Accès refusé |
| 3 | Code correct MAIS âge < 18 | Accès interdit aux mineurs |
| 4 | Prénom = `admin` ou `agent007` | Message spécial |

---

## Contraintes techniques

- Au moins une variable pour stocker le code secret
- `input()` pour chaque information demandée
- `int()` pour convertir l'âge
- Au moins un `if`, un `elif`, un `else`
